```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```
camada = [keras.layers.Dense(units=1, input_shape=[1])]
modelo_neural = tf.keras.Sequential(camada)
modelo_neural.compile(optimizer='sgd', loss='mean_squared_error')
```

```
X = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
y = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

```
modelo_neural.fit(X,y,epochs=500)
```

```
print(modelo_neural.predict([10.0]))
```
